```python
#merging and joining data sets
#types -inner,left,right,outer
#use merge() to combine datasets, based on keys
#can merge 2 dataframs on the same column, in our case this is product ID

sales_df = pd.read_csv('/content/sample_data/sales_data.csv')
products_df = pd.read_csv('/content/sample_data/product_customer_data.csv')
sales_df.head()
products_df.head()


merged_df = pd.merge(
    sales_df,
    products_df,
    on='ProductID',
    how='inner'


)
merged_df.head()
```

```python
# merge on ProductID
merged_df = pd.merge(
    sales_df,
    products_df,
    on="ProductID",
    how="inner"
)

merged_df.head()
```

```python
#pivot tables
#a way to summarise data in matrix format
#useful for multi-dimentional analysis (rows vs columns vs values)


#1.read in our csv files
#2.merge(done above)
#3. create piot table -summarise total sales by region and product category

sales_df = pd.read_csv('/content/sample_data/sales_data.csv')
products_df = pd.read_csv('/content/sample_data/product_customer_data.csv')


#create new column  for toatal sales

merged_df ['TotalSales']  = merged_df['TotalSaleAmount'] * merged_df['UnitPrice']

#pivot table-total sales by region and product category

pivot_table= pd.pivot_table(
    merged_df,
    index='CustomerCountry',
    columns = 'ProductCategory',
    values ='TotalSales',
    aggfunc='sum',
    fill_value=0



)
print(pivot_table)

```
