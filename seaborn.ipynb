```python
#exploring seaborn default data sets

tips = sns.load_dataset('tips')
titanic =sns.load_dataset('titanic')
tips
```

```python
#basic plots
#scatter plot
sns.scatterplot(x='tip',y='total_bill', data = tips, hue='day', size='size' )
```

```python
sns.scatterplot(x='size',y='total_bill', data = tips)
```

```python
sns.scatterplot(x='sex',y='total_bill', data = tips)
```

```python
sns.scatterplot(x='size',y='tip', data = tips, hue='day')
```

```python
#basic plots
#scatter plot
sns.scatterplot(x='tip',y='total_bill', data = tips, hue='day', size='size' ,palette='YlGnBu')
```

```python
#histplot

#what it is and why/when it is usefult
#histogram =groups data into bins and show frequency distribution
#why/when-understand distribution and spread of single variable

sns.histplot(tips['tip'], kde=True, bins=20) # kde trendline
```

```python
#bar plot
#chart that represents categorical gata with rectangular bar-height reflects value
#useful for comparison

sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
#boxplot
#summarises the distribution of a dataset, showing median quartiles and outliers
#good for comparing distributions.

sns.boxplot(x='day', y='tip', data=tips, hue='sex' , palette='YlGnBu')
```

```python
#stripplot
#displays individual data points, often layered with a box plot for detail
#adds bdetail showing the spread of actual observations

sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette='YlGnBu', dodge='True') #doge spreads out the datapoints
```

```python
sns.jointplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="scatter"
)
```

```python
sns.pairplot(tips, diag_kind="kde",hue="smoker"
)
```

```python
sns.heatmap(
    tips[["total_bill", "tip", "size"]].corr(),
    annot=True,
    cmap="magma"
)
```
